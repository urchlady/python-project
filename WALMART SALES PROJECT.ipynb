```python
#Explore the relationship between "Date" and "Weekly_Sales" columns to understand 
#the sales performance over time. Identify trends, seasonality patterns, and any significant changes in sales.
# Convert 'Date' column to datetime type with the correct format
data['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')

# Set 'Date' column as the index
data.set_index('Date', inplace=True)

# Resample the data to a weekly frequency and calculate the sum of weekly sales
weekly_sales = df['Weekly_Sales'].resample('W').sum()

# Plot the weekly sales over time
plt.figure(figsize=(12, 6))
plt.plot(weekly_sales)
plt.title('Weekly Sales Performance')
plt.xlabel('Date')
plt.ylabel('Weekly Sales')
plt.show()
```

```python
import plotly.express as px

# Create a line chart using Plotly Express
fig = px.line(weekly_sales, title='Weekly Sales Performance')

# Display the chart
fig.show()
```

```python
#Investigate the impact of holidays on sales by analyzing the "Holiday_Flag" column. 
# Compare sales during holidays versus non-holiday periods to determine if there is a significant difference 
# and if holidays drive higher sales.


# Filter the data for holiday and non-holiday periods
holiday_sales = df[df['Holiday_Flag'] == 1]['Weekly_Sales']
non_holiday_sales = df[df['Holiday_Flag'] == 0]['Weekly_Sales']

# Calculate the average sales during holiday and non-holiday periods
avg_sales_holiday = holiday_sales.mean().round()
avg_sales_non_holiday = non_holiday_sales.mean().round()

# Print the average sales during holiday and non-holiday periods
print("Average Sales during Holidays: ", avg_sales_holiday)
print("Average Sales during Non-Holidays: ", avg_sales_non_holiday)
```

```python
# Examine the relationship between "Temperature" and "Weekly_Sales" columns to assess the impact of temperature on sales. 
# Analyze if there is a correlation or if certain temperature ranges influence consumer buying behavior.

import numpy as np

```

```python
def calculate_Temperature_distribution(Temperature):
    if Temperature < 47:
        return '42-46'
    elif Temperature < 53:
        return '47-52'
    elif Temperature < 59:
        return '53-58'
    elif Temperature < 65:
        return '59-64'
    elif Temperature < 71:
        return '64-70'
    else:
        return '71 and over'

df['temperature_bins'] = df['Temperature'].apply(calculate_Temperature_distribution)
```

```python
df
```

```python
wt=df.groupby('temperature_bins')['Weekly_Sales'].mean().round()
```

```python
wt
```

```python
plt.bar(wt.index, wt.values)
plt.xlabel('temperature_bins')
plt.ylabel('Weekly_Sales')
plt.title('Average Weekly Sales by Temperature Range')

plt.show()
```

```python
# Calculate the correlation between 'Temperature' and 'Weekly_Sales'
correlation = data['Temperature'].corr(data['Weekly_Sales'])

# Plot the scatter plot of 'Temperature' vs. 'Weekly_Sales'
plt.figure(figsize=(10, 6))
plt.scatter(data['Temperature'], data['Weekly_Sales'], c=data['Temperature'], cmap='coolwarm')
plt.title('Temperature vs. Weekly Sales')
plt.xlabel('Temperature')
plt.ylabel('Weekly Sales')
plt.show()

# Print the correlation value
print("Correlation between Temperature and Weekly Sales: ", correlation)
```

```python
# Investigate the relationship between "Fuel_Price" and "Weekly_Sales" columns to
# understand the impact of fuel prices on sales. Assess if higher fuel prices affect consumer spending patterns.

# Define the fuel price ranges
fuel_price_ranges = ['2.5-3', '3-3.5', '3.5-4']

# Categorize the fuel prices into ranges
df['Fuel_Price_Range'] = pd.cut(df['Fuel_Price'], bins=[2.5, 3, 3.5, float('inf')], labels=fuel_price_ranges)

# Calculate the average weekly sales for each fuel price range
average_sales_by_range = df.groupby('Fuel_Price_Range')['Weekly_Sales'].mean()

# Plot the bar chart
plt.figure(figsize=(10, 6))
average_sales_by_range.plot(kind='bar')
plt.title('Average Weekly Sales by Fuel Price Range')
plt.xlabel('Fuel Price Range')
plt.ylabel('Average Weekly Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose


# Assuming your dataset is stored in a DataFrame called 'df' with columns
result_list = []

for column in df1.columns:
    result = seasonal_decompose(df1[column], model='additive', period=36)
    result_list.append(result)

# You can access the decomposed components for each column using result_list[i].trend, 
# result_list[i].seasonal, and result_list[i].resid
```

```python
result.plot()
plt.xticks(rotation=45)
plt.show()
```

```python
# testing for stationarity of df1 using adfuller
from statsmodels.tsa.stattools import adfuller

def test_stationarity(timeseries):
    #perform Dickey-Fuller test
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag= 'AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used'])
    print(dfoutput)
test_stationarity(df['Weekly_Sales'])
```

```python
# fitting ts into SARIMAX model
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
model = SARIMAX(df['Weekly_Sales'], order=(2,2,2), seasonal_order=(2,2,2,12))
model_fit = model.fit()
```

```python
df1
```

```python
# Forecast for the next one year
forecast_steps = 12
forecast = model_fit.forecast(steps=forecast_steps)
```

```python
forecast
```

```python
# Generate dates for the forecast period
forecast_index = pd.date_range(start=df.index[-1], periods=forecast_steps + 1, freq='M')[1:]
forecast_index
```

```python
# Plot the original data and forecast
plt.figure(figsize=(10, 6))
plt.plot(df1.index, df1['Weekly_Sales'], color='blue', label='Original Data')
plt.plot(forecast_index, forecast, color='red', label='Forecast')
plt.title('Weekly Forecast For The Next 1 Year')
plt.xlabel('Date')
plt.ylabel('Weekly_Sales')
plt.legend()
plt.grid(True)
plt.show()
```
